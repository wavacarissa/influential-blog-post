```python
import pandas as pd
import datetime
```

### Load the Data

```python
source = 'C:\\Users\\wavac\\Bangkit\\Machine Learning Assignment\\Dataset\\posts.csv'

posts = pd.read_csv(source, index_col=0,  names=['post_id','title','blogger_name','blogger_id','num_of_comment','content','url','date','retrieved_links','retrieved_comment','num_words_s','num_words','avg_word_len_s',
                           'avg_word_len','MEIBI','MEIBIX'])
```

```python
posts.head()
```

### Check for empty/null data

```python
null_columns=posts.columns[posts.isnull().any()]
posts[null_columns].isnull().sum()
```

Since we're not going to use 'Title' and 'Content' for our analysis, we could drop the attributes.

### Strip datetime into individual year, month and day
<br> Since our model cannot handle data in the form of datetime

```python
date = posts['date'].str.split('-',expand=True)
date.columns = ['year','month','day']
posts = posts.join(date)
```

### Drop unecessary attributes

```python
dropped_columns = ['title','content','url','date']
posts.drop(dropped_columns, axis=1, inplace=True)
```
