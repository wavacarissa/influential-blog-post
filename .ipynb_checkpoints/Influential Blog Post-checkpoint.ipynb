### Removing redundant features

We're going to use correlation matrix to decide which features that were important for this analysis. Correlation matrix is a matrix that was consisted of correlation value (pearson's correlation) between every combination of attributes. The higher the number is, the higher the correlation, and the more linearly dependent the data is. We're going to remove features that linearly dependent to other features.

```python
fig, ax = plt.subplots(figsize=(17,15)) 
matrix = np.triu(clean_posts.corr())
sns.heatmap(clean_posts.corr(), annot=True)
plt.show()
```

In this analysis, we're going to remove several features, such as: <br>
- Number of words (Since it has a high correlation (1) towards the Number of Words (with stop words))
- Average word length (Since it has a high correlation (1) towards the Average word length (with stop words))
- Retrieved comment (Since it has a high correlation (0.71) towards the Number of Comments)
- Retrieved Links (Since it has a high correlation (0.71) towards the Number of Comments)

Since the features had a correlation of >0.5. We decided to take features that the correlation value was <0.5.



```python
dropped_columns = ['num_words','avg_word_len','retrieved_comment','retrieved_links']
clean_posts.drop(dropped_columns, axis=1, inplace=True)
```

Here's how the dataset looks after removing redundant features

```python
fig, ax = plt.subplots(figsize=(17,15)) 
matrix = np.triu(clean_posts.corr())
sns.heatmap(clean_posts.corr(), annot=True)
plt.show()
```

# Splitting Training, Validation, and Testing Data<br>
We're going to split the dataset into training and testing data. We're going to use the feature MEIBI as our target value.


```python

train= clean_posts.sample(frac=0.8,random_state=0)
test = clean_posts.drop(train.index)
Y_train = train['MEIBI']
Y_test = test['MEIBI']
```

### Normalization

In this section, we decided to remove the datetime attribute (year,month,and day feature). Normalization is applied to the remaining features so in that it could made the training process easier, since the values were scaled into values between 0 and 1.

```python
datetime_columns = ['month','year','day']
target_column = ['MEIBI']
datetime_columns.extend(target_column)
dropped = datetime_columns
train_stats = train.describe().transpose()
```

```python
def norm(x):
    return (x - train_stats['mean']) / train_stats['std']

normed_train = norm(train)
normed_test = norm(test)

normed_train.drop(dropped, axis=1, inplace=True)
normed_test.drop(dropped, axis=1, inplace=True)
```

```python
normed_train.head()
```

### Removing Outliers<br>
Before removing outliers, we had to take a look into the distribution for each features in the data.

```python
sns.pairplot(normed_train, diag_kind="kde")
plt.show()
```

We can see that most of the data are leaning towards the left side, implying that there were several outliers in the right side of the diagram. We're going to remove the outliers, which we defined as values with a z-score >=3

```python
from scipy import stats

removed_outliers = normed_train[(np.abs(stats.zscore(normed_train)) < 3).all(axis=1)]
sns.pairplot(removed_outliers, diag_kind="kde")
plt.show()
```

### Final Data

```python
X_train = removed_outliers
Y_train = Y_train[X_train.index]
X_val = X_train.sample(frac=0.2,random_state=0)
Y_val = Y_train[X_val.index]
X_test = normed_test
Y_test = Y_test
```

```python
X_train.to_csv('data/X_train.csv')
Y_train.to_csv('data/Y_train.csv')
X_test.to_csv('data/X_test.csv')
Y_test.to_csv('data/Y_test.csv')
```

# Model <br>
We decided to use multilayer perceptron (Neural Network) as our model, since they were known for their capabilities to solve non-linear problems. We also applied Keras Tuner in order to tune the model's hyperparameters.

```python
import tensorflow as tf
from tensorflow import keras as keras
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.models import Sequential
from kerastuner.applications import HyperResNet
from kerastuner.tuners import Hyperband
from kerastuner.tuners import RandomSearch
from kerastuner.engine.hyperparameters import HyperParameter

import time
import os


if os.path.isdir('logs') == False:
    os.mkdir('logs')
    
LOG_DIR = f"{int(time.time())}"
```

In this section, we're going to define our model in build_model(). We used Dense layers and ReLU as our activation function(s). Since we are using keras tuner to help us tuning the model's hyperparameters, we could see several hp.Int(......) within the parameter. This is keras tuner's way to find a good combination of parameters for our model, by placing values within the range specified in the hp.Int(.....)

```python
def build_model(hp):
    model = keras.models.Sequential()
    
    #Input Layer
    model.add(Dense(hp.Int('input_units',min_value=32,max_value=256,step=32),input_dim=len(X_train.keys())))
    model.add(Activation('relu'))
    
    #n Hidden Layer
    for i in range(hp.Int('n_layers', 1, 4)):
        hidden_units = 'hidden_'+str(i)+'_units'
        model.add(Dense(hp.Int(hidden_units,
                               min_value=32,
                               max_value=256,
                               step=32)))
        model.add(Activation('relu'))
        
    #Output Layer
    model.add(Dense(1))
    model.add(Activation('relu'))
              

#     model = keras.Sequential([
#         layers.Dense(hp.Int('input_units',min_value=32,max_value=256,step=32), activation='relu', input_shape=[len(X_train.keys())]),
#         layers.Dense(64, activation='relu'),
#         layers.Dense(1)
#       ])

    model.compile(loss='mse',
                    optimizer="adam",
                    metrics=['accuracy'])
    return model
```

```python
tuner = RandomSearch(
        build_model,
        objective = 'val_accuracy',
        max_trials = 3,
        executions_per_trial = 3,
        directory = LOG_DIR)
```

```python
tuner.search_space_summary()
```

```python
tuner.search(x=X_train, y=Y_train,epochs=100,batch_size=64,validation_data=(X_val,Y_val),verbose=0)
```

### Best hyperparameters according to Keras Tuner <br>
It is shown that with an input of 128 nodes, and 2 hidden layers (layer 0 with 192 nodes), the model could produce prediction with a high accuracy.
