```python
import pandas as pd
import datetime
```

### Load the Data

```python
source = 'data/posts.csv'

posts = pd.read_csv(source, index_col=0,  names=['post_id','title','blogger_name','blogger_id','num_of_comment','content','url','date','retrieved_links','retrieved_comment','num_words_s','num_words','avg_word_len_s',
                           'avg_word_len','MEIBI','MEIBIX'])
```

```python
posts.head()
```

### Check for empty/null data

```python
null_columns=posts.columns[posts.isnull().any()]
posts[null_columns].isnull().sum()
```

Since we're not going to use 'Title' and 'Content' for our analysis, we could drop the attributes.

### Strip datetime into individual year, month and day
<br> Since our model cannot handle data in the form of datetime

```python
date = posts['date'].str.split('-',expand=True)
date.columns = ['year','month','day']
date = date.applymap(lambda x: int(x))
posts = posts.join(date)
```

### Drop unecessary attributes

```python
dropped_columns = ['title','content','url','date','blogger_name','blogger_id','MEIBIX']
posts.drop(dropped_columns, axis=1, inplace=True)
```

### Preprocessed data

```python
posts.head(10)
```

```python
clean_posts = posts
clean_posts.to_csv('data/clean_posts.csv')
```

### Removing redundant features

We're going to use correlation matrix to decide which features that were important for this analysis. Correlation matrix is a matrix that was consisted of correlation value (pearson's correlation) between every combination of attributes. The higher the number is, the higher the correlation, and the more linearly dependent the data is. We're going to remove features that linearly dependent to other features.

```python
fig, ax = plt.subplots(figsize=(17,15)) 
matrix = np.triu(clean_posts.corr())
sns.heatmap(clean_posts.corr(), annot=True)
plt.show()
```

In this analysis, we're going to remove several features, such as: <br>
- Number of words (Since it has a high correlation (1) towards the Number of Words (with stop words))
- Average word length (Since it has a high correlation (1) towards the Average word length (with stop words))
- Retrieved comment (Since it has a high correlation (0.71) towards the Number of Comments)
- Retrieved Links (Since it has a high correlation (0.71) towards the Number of Comments)

Since the features had a correlation of >0.5. We decided to take features that the correlation value was <0.5.



```python
dropped_columns = ['num_words','avg_word_len','retrieved_comment','retrieved_links']
clean_posts.drop(dropped_columns, axis=1, inplace=True)
```

Here's how the dataset looks after removing redundant features

```python
fig, ax = plt.subplots(figsize=(17,15)) 
matrix = np.triu(clean_posts.corr())
sns.heatmap(clean_posts.corr(), annot=True)
plt.show()
```
